```python
def open_api():
    
    
#importing libraries
    from requests.exceptions import ConnectionError
    import pandas as pd
    import requests
    import json
    
    
#input of API url
    url = input('enter API URL')
    
    
#try except for checking if the url is valid
    try:
        
#requesting data from API, printing response, reading with json handler, printing the dictornary keys
        r = requests.get(url, timeout=10)
        print(r)
        json = r.json()
        parameters= json.keys()
        print ('The paramters available in API are',  '\n', parameters)
        
#input of which parameter data is to be fetched        
        paramter_open= input('enter parameter to open')
        
#checking if the parameter entered is available in the key or not        
        if paramter_open not in parameters:
            print ('please enter valid parameter from keys')

#converting the data from json to dataframe to be understanable easiely
        else:
            #print(json[key])
            df = pd.DataFrame(json[paramter_open])
            return df
        
    except ConnectionError:
        print ('Failed to open API URL. Please enter correct API URL')
```

```python
open_api()
```
