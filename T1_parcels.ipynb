```python
#importing libraries
import pandas as pd
import requests
```

```python
#saving API url as variable 
url = 'https://demodev2.kappazeta.ee/ard_api_demo/v1/parcel/?out_geom_fmt=GeoJSON&in_geom_fmt=WKT&properties=parcel_id%2Cparcel_foreign_id%2Cclient_foreign_id%2Ccrop_foreign_code%2Clanduse_foreign_code%2Carea%2Cgeometry'
```

```python
#requesting data to be fetched from url 
r = requests.get(url)
```

```python
print(r)
```

```python
#reading data from json handler
json = r.json()
```

```python
#checking dictornary keys in which data is stored
json.keys()
```

```python
json['parcels']
```

```python
#converting data from json to dataframe so it can be understood easily
df = pd.DataFrame(json['parcels'])
```
