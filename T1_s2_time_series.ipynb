```python
#importing libraries
import pandas as pd
import requests
```

```python
#saving API url as variable 
url = 'https://demodev2.kappazeta.ee/ard_api_demo/v1/time_series/s2?limit_to_rasters=true&parcel_foreign_id=1&properties=parcel_foreign_id%2Cs2product_start_time%2Cs2product_ron%2Cndvi_med'
```

```python
#requesting data to be fetched from url 
r = requests.get(url)
```

```python
print(r)
```

```python
#reading data from json handler
json = r.json()
```

```python
#checking dictornary keys in which data is stored
json.keys()
```

```python
json['s2_time_series']
```

```python
#converting data from json to dataframe so it can be understood easily
df = pd.DataFrame(json['s2_time_series'])
```
