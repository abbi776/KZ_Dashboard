```python
#importing libraries
import pandas as pd
import requests
```

```python
#saving API url as variable
url = 'https://demodev2.kappazeta.ee/ard_api_demo/v1/time_series/s1?limit_to_rasters=true&parcel_foreign_id=1&properties=parcel_foreign_id%2Cs1product_end_time%2Cs1product_ron%2Ccohvh_med%2Ccohvv_med%2Cvhvv_med'
```

```python
#requesting data to be fetched from url
r = requests.get(url)
```

```python
print(r)
```

```python
#reading data from json handler
json = r.json()
```

```python
#checking dictornary keys in which data is stored
json.keys()
```
