```python
#importing libraries
import pandas as pd
import requests
```

```python
#saving API url as variable
url = 'https://demodev2.kappazeta.ee/ard_api_demo/v1/raster_index/?raster_types=s0%2Ccoh6%2Ccoh12%2Csndvi%2Cndvi&out_geom_fmt=GeoJSON&in_geom_fmt=WKT&round_to_day=true&properties=raster_type%2Cdatetime%2Cron'
```

```python
#requesting data to be fetched from url
r = requests.get(url)
```

```python
print(r)
```

```python
#reading data from json handler
json = r.json()
```

```python
#checking dictornary keys in which data is stored
json.keys()
```

```python
json['raster_index']
```

```python
#converting data from json to dataframe so it can be understood easily
df = pd.DataFrame(json['raster_index'])
```
